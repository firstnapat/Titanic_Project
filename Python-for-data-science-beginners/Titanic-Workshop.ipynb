### Excercise

คำถาม : หาสัดส่วนของผู้โดยสารแยกตามคลาส

```python
cl1 = df['Pclass'].value_counts()[1]
cl2 = df['Pclass'].value_counts()[2]
cl3 = df['Pclass'].value_counts()[3]
csum = df['Pclass'].value_counts().sum()
```

```python
print('% of class 1:'+str(cl1/csum))
print('% of class 2:'+str(cl2/csum))
print('% of class 3:'+str(cl3/csum))
```

คำถาม : หาว่าผู้โดยสารชั้น 1 2 3 รอดชีวิตเป็นกี่เปอร์เซ็นต์ของผู้โดยสารชั้นนั้น ๆ ทั้งหมด

```python
sur1 = df[df['Pclass']==1]['Survived'].value_counts()[1]
sur2 = df[df['Pclass']==2]['Survived'].value_counts()[1]
sur3 = df[df['Pclass']==3]['Survived'].value_counts()[1]
```

```python
sursum1 = df[df['Pclass']==1]['Survived'].value_counts().sum()
sursum2 = df[df['Pclass']==2]['Survived'].value_counts().sum()
sursum3 = df[df['Pclass']==3]['Survived'].value_counts().sum()
```

```python
print('% of sur class 1:'+str(sur1/sursum1))
print('% of sur class 2:'+str(sur2/sursum2))
print('% of sur class 3:'+str(sur3/sursum3))
```
