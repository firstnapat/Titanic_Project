### Exercise

เติม na อายุของผู้หญิงคลาส 3 ด้วยค่ากลาง(Median)

เติม na อายุของผู้ชายคลาส 3 ด้วยค่าเฉลี่ย(Mean)

```python
median_fc3 = df_copy[(df_copy['Pclass']==3) & (df_copy['Sex']=='female')]['Age'].median()
median_fc3
```

```python
# เติม na ของผู้หญิงคลาส3
df_copy.loc[(df_copy['Pclass']==3) & (df_copy['Sex']=='female'),'Age'] = df_copy[(df_copy['Pclass']==3) & 
                                                                                 (df_copy['Sex']=='female')]['Age'].fillna(median_fc3)
```

```python
df_copy[(df_copy['Pclass']==3) & (df_copy['Sex']=='female')]['Age'].value_counts()
```

```python
mean_mc3 = df_copy[(df_copy['Pclass']==3) & (df_copy['Sex']=='male')]['Age'].mean()
mean_mc3
```

```python
# เติม na ของผู้ชายคลาส3
df_copy.loc[(df_copy['Pclass']==3) & (df_copy['Sex']=='male'),'Age'] = df_copy[(df_copy['Pclass']==3) & 
                                                                                 (df_copy['Sex']=='male')]['Age'].fillna(mean_mc3)
```

```python
df_copy[(df_copy['Pclass']==3) & (df_copy['Sex']=='male')]['Age'].value_counts()
```
